```python
import numpy as np

# read the file contents into a list
with open('outt.txt') as f:
    lines = f.readlines()
# remove any leading/trailing white spaces and convert each line to a list of characters
lines = [list(line.split()) for line in lines[:-1]]

# convert the list of lists to a NumPy array
arr = np.array(lines)

# convert the character array to a numeric array, replacing anyon-numeric characters with zeros
arr = arr.astype(float)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Fixing random state for reproducibility
np.random.seed(19680801)


def randrange(n, vmin, vmax):
    """
    Helper function to make an array of random numbers having shape (n, )
    with each number distributed Uniform(vmin, vmax).
    """
    return (vmax - vmin)*np.random.rand(n) + vmin

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

n = len(arr)

# For each set of style and range settings, plot n random points in the box
# defined by x in [23, 32], y in [0, 100], z in [zlow, zhigh].
for m, zlow, zhigh in [('o', -50, -25), ('^', -30, -5)]:
    xs = [i[0] for i in arr]
    ys =[i[1] for i in arr]
    zs = [i[2] for i in arr]
    ax.scatter(xs, ys, zs, marker=m)

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```
